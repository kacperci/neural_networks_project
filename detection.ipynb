```python
from pathlib import Path
from ultralytics import YOLO

DATA_ROOT = Path('Dataset_BUSI_YOLO')
DATA_YAML = DATA_ROOT / 'data.yaml'

if not DATA_YAML.exists():
    raise FileNotFoundError(f'Brak {DATA_YAML}. Najpierw uruchom prepare_data.ipynb.')

print('Korzystam z konfiguracji:', DATA_YAML.resolve())
```

```python
MODEL_WEIGHTS = 'yolo11n.pt'
# Alternatywa: 'yolov8n.pt'
IMG_SIZE = 640
EPOCHS = 50
BATCH = 16
PROJECT = 'runs_busi_detection'
EXPERIMENT_NAME = 'yolo11n_busi'

model = YOLO(MODEL_WEIGHTS)
model
```

```python
train_results = model.train(
    data=str(DATA_YAML),
    imgsz=IMG_SIZE,
    epochs=EPOCHS,
    batch=BATCH,
    project=PROJECT,
    name=EXPERIMENT_NAME,
    exist_ok=True
)
train_results
```

```python
metrics = model.val(data=str(DATA_YAML), imgsz=IMG_SIZE)
metrics
```

```python
from itertools import islice

val_images_dir = DATA_ROOT / 'images' / 'val'
sample_images = list(islice(sorted(val_images_dir.glob('*.*')), 6))

if not sample_images:
    raise FileNotFoundError(f'Brak obrazow walidacyjnych w: {val_images_dir}')

pred_results = model.predict(source=[str(p) for p in sample_images], conf=0.2, save=True)
pred_results
```
